```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
def occurance_count(a_df):
    occurance = Counter()
    for entry in a_df:
        occurance[entry] = occurance[entry] + 1
    return occurance
```

```python
orders_df = pd.read_csv("C:~/orders.csv")
```

```python
orders_df.shape
```

```python
tmp_dict = occurance_count(orders_df['user_id'])
```

```python
tmp_dict_df = pd.DataFrame.from_dict(tmp_dict, orient='index').reset_index()
print(tmp_dict_df.head())
```

```python
plt_value_usage = occurance_count(tmp_dict_df[0])
print(plt_value_usage)
```

```python
plot_lists = sorted(plt_value_usage.items()) # sorted by key, return a list of tuples

x, y = zip(*plot_lists) # unpack a list of pairs into two tuples

plt.figure(figsize=(16,8))
sns.set_style("whitegrid")
sns.barplot(x, y, color='olive')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Maximum order number', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
products_df = pd.read_csv("C:~/instacart_2017_05_01/products.csv")
```

```python
cnt_prod_id_df1 = pd.merge(cnt_prod_id_df, products_df, on="product_id", how="left")
```

```python
cnt_prod_id_df.head(40)
```

```python
cnt_prod_id_df1 = cnt_prod_id_df[0:40]
cnt_prod_id_df1 = cnt_prod_id_df1.sort_values('count').reset_index().iloc[:,[1,2,3,4,5]]
```
